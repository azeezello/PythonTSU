# Рекурсия
## Задача 1: «Сумма чисел»

**Описание**: Написать рекурсивную функцию, которая вычисляет сумму чисел от 1 до N.

**Что нужно сделать**:

*   Создать функцию recursive_sum() с параметром n
*   Реализовать базовый случай (когда n = 1)
*   Реализовать рекурсивный случай (n + сумма от 1 до n-1)
*   Протестировать функцию на разных значениях

**Пример работы программы**:



```
РЕКУРСИВНАЯ СУММА ЧИСЕЛ

Введите число N: 5

Сумма чисел от 1 до 5 = 15
```






````python
def recursive_sum(n):

    if n == 1:
        return n
    else:
        return(n + recursive_sum(n-1))


print("РЕКУРСИВНАЯ СУММА ЧИСЕЛ\n")

N_input = int(input("Введите число N: "))

print(f"Сумма чисел от 1 до {N_input} = {int(recursive_sum(N_input))}")



````

## Задача 2: «Палиндром-детектор»
**Описание**: Напишите рекурсивную функцию, которая проверяет, является ли строка палиндромом (читается одинаково слева направо и справа налево). Игнорируйте регистр букв и пробелы.

**Что нужно сделать**:

*   Создать рекурсивную функцию is_palindrome(text)
*   Функция должна игнорировать регистр и пробелы
*   Использовать базовый случай для пустой строки или строки из одного символа
*   Вывести результат проверки для нескольких тестовых строк

**Входные данные**: строка

**Выходные данные**: логическое значение (True/False)

**Пример работы программы**:

```
ПАЛИНДРОМ-ДЕТЕКТОР

=== ТЕСТИРОВАНИЕ ФУНКЦИИ ===
"радар" – True
"Python" – False  
"А роза упала на лапу Азора" – True
"мадам" – True
"программирование" – False
```




````python
def is_palindrome(text):
    text_clean = text.lower().replace(" ","")
    if len(text_clean) <= 1:
        return True
    elif text_clean[0] != text_clean[-1]:
        return False
    else:
        return is_palindrome(text_clean[1:-1])

print("ПАЛИНДРОМ-ДЕТЕКТОР\n")

print("=== ТЕСТИРОВАНИЕ ФУНКЦИИ ===")

text_input = input()

print(f"{is_palindrome(text_input)}")
````

# Декораторы
## Задача 1: «Декоратор-приветствие»

**Описание**: Создать простой декоратор, который добавляет приветственное сообщение перед выполнением функции.

**Что нужно сделать**:

*   Создать декоратор greeting_decorator
*   Декоратор должен выводить “Функция запускается…” перед вызовом функции
*   Выводить “Функция завершена!” после выполнения функции
*   Применить к двум простым функциям (на Ваше усмотрение)

**Входные данные**: любые функции

**Выходные данные**: приветственные сообщения

**Пример работы программы**:



```
ДЕКОРАТОР-ПРИВЕТСТВИЕ

Функция запускается...
Привет, мир!
Функция завершена!

Функция запускается...
Результат: 15
Функция завершена!

Все функции выполнены!
```





````python
def greeting_decorator(func):
    def wrapper():
        print("Функция запускается...")
        func()
        print("Функция завершена!\n")
    return wrapper

@greeting_decorator
def greet():
    print("Привет, мир")


@greeting_decorator
def result():
    print("Результат: 15")


greet()
result()
print("Все функции выполнены!")


````

Для тестирования используйте функции:



```

    return a * b

@positive_numbers_only
```

````python
def positive_numbers_only (func):
    def wrapper(*args, **kwargs):
        for arg in args:
            if arg < 0:
                print(f"Аргумент {arg} не является положительным числом!")
            return "Результат: " func()

@positive_numbers_only
def multiply(a,b):
    return a * b
````
