 ## Задача 1: "Текстовая статистика"

**Описание**: Создайте простую программу, которая поможет пользователю посчитать количество символов и слов, а также количество предложений, введенных пользователем с клавиатуры.

```python
print(str.upper("анализатор текстовой статистики"))
text = input('Введите текст для анализа: ')
words = len(text.split())
sentences = text.count('.') + text.count('!') + text.count('?')
#Пример текста для проверки: Добрый день! Как дела? Все хорошо.

print()
print(str.upper('=== cтатистика текста ==='))

print("Общее количество символов: ", len(text))
print(f"Количество слов: {words}")
print(f"Количество предложений: {sentences}")
```

## Задача 2: "Планировщик бюджета путешествия"

**Описание**: Помогите пользователю рассчитать общий бюджет для путешествия, учитывая все основные пункты расходов.

```python
print(str.upper("планировщик бюджета путешествия"))

flight = int(input("Введите бюджет на перелет: "))
hotel = int(input("Введите бюджет на отель: "))
food = int(input("Введите бюджет на питание в день: "))
days = int(input("Введите количество дней: "))
entertainment = int(input("Введите бюджет на развлечения: "))

food_total = food * days
budget_total = flight + hotel + food_total + entertainment

print()
print(str.upper("=== финансовый план путешествия ==="))
print(f"Перелет: {flight} руб")
print(f"Проживание: {hotel} руб")
print(f"Питание: {food_total} руб ({days} дней х {food} руб/день)")
print(f"Развлечения: {entertainment} руб")
print("-"*20)
print(f"ОБЩИЙ БЮДЖЕТ: {budget_total} руб")

print()
print("Хорошего путешествия!")

```

## Задача 1: "Плейлист любимых песен"

**Описание**: Создайте плейлист из любимых песен и научитесь работать с основными операциями над списками.

```python
playlist = []

print(str.upper("создание плейлиста любимых песен"))

print()

print("Добавьте 5 любимых песен:")
# Bohemian Rhapsody
# Hotel California
# Imagine
# Yesterday
# Smells Like Teen Spirit

playlist = []

print(str.upper("создание плейлиста любимых песен"))

print()

print("Добавьте 5 любимых песен:")

# добавляем песни в плейлист
for i in range(5):
    song = input(f"Песня {i+1}: ")
    playlist.append(song)

# возвращаем созданный плейлист
print()
print(str.upper("=== ваш плейлист ==="))

for s in range(len(playlist)):
    print(f"{s+1}. {playlist[s]}")

print()
print("Первая песня: ", playlist[0])
print("Последняя песня: ", playlist[4])
print("Всего песен: ", len(playlist))

print()
print("Удаляем песню №3: ", playlist[2])
playlist.remove(playlist[2])

print()
print(str.upper("=== обновленный плейлист ==="))
for s in range(len(playlist)):
    print(f"{s+1}. {playlist[s]}")
```

## Задача 2: "Топ-5 фильмов с рейтингом"

**Описание**: Создайте рейтинг топ-5 фильмов с оценками и научитесь извлекать данные по позициям

```python
print(str.upper("топ-5 фильмов с рейтингом"))
print()
print("Введите 5 лучших фильмов:")
movies = []

# добавляем фильмы в список
for i in range(5):
  movie = input(f"Фильм {i+1}: ")
  movies.append(movie)

print()

# вводим рейтинги
print("Введите рейтинги (1-10):")

ratings = []

for i in range(5):
  rating = float(input(f"Рейтинг{i+1}: "))
  ratings.append(rating)

print()

# возвращаем рейтинг фильмов
print(str.upper("=== рейтинг фильмов ==="))
for i in range(len(movies)):
  print(f"{i+1}. {movies[i]} - {ratings[i]}/10")

print()
print(f"Первый фильм: {movies[0]} ({ratings[0]}/10)")
print(f"Последний фильм: {movies[4]} ({ratings[4]}/10)")
print(f"Средний рейтинг: {sum(ratings)/len(ratings)}")
```

# Множества (set)

## Задача 1: "Общие друзья в ВК"

**Описание**: Определите общих друзей между двумя пользователями ВК

```python
print(str.upper("поиск общих друзей в соцсетях"))
print()

user1_friends = input("Введите друзей пользователя 1 (через запятую): ")
user1_set = set(user1_friends.split(','))
user2_friends = input("Введите друзей пользователя 2 (через запятую): ")
user2_set = set(user2_friends.split(','))
сommon_friends = user1_set.intersection(user2_set)


print()
print(str.upper("=== результаты анализа ==="))
print(f"Друзья пользователя 1: {len(user1_set)} человек")
print(f"Друзья пользователя 2: {len(user2_set)} человек")

print("Общие друзья:", ", ".join(сommon_friends))
print(f"Количество общих друзей: {len(сommon_friends)}")
```

## Задача 2: "Анализ покупок в супермаркете"

**Описание**: Проанализируйте покупки двух покупателей и найди уникальные и общие товары

```python
print(str.upper("анализ покупок в супермаркете"))

print()
items1 = input("Введите товары покупателя 1 (через запятую): ") # молоко,хлеб,яйца,сыр,колбаса
buyer1_items = set(items1.split(','))
items2 = input("Введите товары покупателя 2 (через запятую): ") # хлеб,сыр,вода,фрукты,конфеты
buyer2_items = set(items2.split(','))

unique_items = buyer1_items.union(buyer2_items)
dif_items1 = buyer1_items.difference(buyer2_items)
dif_items2 = buyer2_items.difference(buyer1_items)
common_items = buyer1_items.intersection(buyer2_items)

print()
print(str.upper("=== результаты анализа ==="))
print("Все уникальные товары:", ", ".join(unique_items))

print()
print("Только у покупателя 1:", ", ".join(dif_items1))
print("Только у покупателя 2:", ", ".join(dif_items2))
print("Общие покупки:", ", ".join(common_items))

print()
print(f"Всего уникальных товаров: {len(unique_items)}")
```
